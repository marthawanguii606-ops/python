```python
average_age = df['age'].mean()
df['age_filled'] = df['age'].fillna(average_age)
```

```python
df['age_filled'].isnull().sum()
```

```python
adults = df[df['age_filled'] > 18]
adults['survived'].mean()
```

```python
kids = df[df['age_filled'] < 18]
kids['survived'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart comparing adult vs kid survival
df['age_group'] = df['age_filled'].apply(lambda x: 'Adult' if x>= 18 else 'Child')
sns.barplot(x='age_group', y='survived', data=df)
plt.title('Survival Rate: Adults vs Children')
plt.ylabel('Survival Rate')
plt.show()
```
